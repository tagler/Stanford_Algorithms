```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# create graph
DG = nx.DiGraph()
```

```python
# add weighted edges 
DG.add_weighted_edges_from([(1,2,1),(1,3,3),(1,4,4),(5,3,5),(4,7,1),
                           (5,6,9),(10,5,4),(1,7,6),(2,3,6),(4,1,3),
                           (4,6,1),(5,11,4),(11,10,2),(7,6,3),(3,9,6),
                           (5,12,3),(12,13,1),(13,14,8),(8,4,3),(14,3,8)])
```

```python
# numer of nodes and edges
print( DG.number_of_nodes() )
print( DG.number_of_edges() )
```

```python
# plot
nx.draw(DG, with_labels=True, font_weight='bold')
```

```python
# find shortest paths
results_path = []
for each_node in DG.nodes():
    try:
        results_path.append( nx.dijkstra_path(DG, 1, each_node, weight='weight') )
    except:
        results_path.append( False )
results_path
```

```python
# find shorted path lengths
results_length = []
for each_node in DG.nodes():
    try:
        results_length.append( nx.dijkstra_path_length(DG, 1, each_node, weight='weight') )
    except:
        results_length.append( False )
results_length
```
