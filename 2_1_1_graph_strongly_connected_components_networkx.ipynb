```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# create graph
DG = nx.DiGraph()
```

```python
# add edges 
DG.add_edges_from([(1,2),(1,3),(1,4),(5,3),(4,7),
                   (5,6),(10,5),(1,7),(2,3),(4,1),
                   (4,6),(5,11),(11,10),(7,6),(3,9),
                   (5,12),(12,13),(13,14),(8,4),(14,3)])
```

```python
# numer of nodes and edges
print( DG.number_of_nodes() )
print( DG.number_of_edges() )
```

```python
# plot
nx.draw(DG, with_labels=True, font_weight='bold')
```

```python
# strongly connected componets lengths
SCCs_lengths = [len(c) for c in sorted(nx.strongly_connected_components(DG), key=len)]
SCCs_lengths[::-1][0:5]
```
