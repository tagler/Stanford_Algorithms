```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# create graph
G = nx.Graph()
```

```python
# add weighted edges 
G.add_weighted_edges_from([(1,2,1),(1,3,3),(1,4,4),(5,3,5),(4,7,1),
                           (5,6,9),(10,5,4),(1,7,6),(2,3,6),(4,1,3),
                           (4,6,1),(5,11,4),(11,10,2),(7,6,3),(3,9,6),
                           (5,12,3),(12,13,1),(13,14,8),(8,4,3),(14,3,8)])
```

```python
# plot
nx.draw(G, with_labels=True, font_weight='bold')
```

```python
# minimun spanning tree
mst = nx.minimum_spanning_edges(G)
rolling_sum = 0
for each in mst:
    rolling_sum = rolling_sum + (each[2]['weight'])
```
